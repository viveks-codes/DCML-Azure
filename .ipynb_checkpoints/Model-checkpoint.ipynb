```python
x = df.iloc[:,:8]
```

```python
y = df.iloc[:,8:9]
```

```python
model = LogisticRegression()
```

```python
model.fit(x,y)
```

```python
unseen = np.array([[10,101,76,48,180,32.9,0.171,63]])
model.predict(unseen)
```

```python
filename = 'model.ml'
pickle.dump(model, open(filename, 'wb'))
```
